# Target Processing

```python
df_lc["loan_status"].value_counts()
```

GOOD LOANS (0)
 - Fully Paid
 - Does not meet the credit policy. Status:Fully Paid

BAD LOANS (1)
 - Charged Off
 - Does not meet the credit policy. Status:Charged Off
 - Late (31-120 days)
 - Late (16-30 days)
 - Default

REMOVE ONGOING LOANS
 - Current
 - In Grace Period

```python
valid_status = [
    "Fully Paid",
    "Charged Off",
    "Late (31-120 days)",
    "Late (16-30 days)",
    "Default",
    "Does not meet the credit policy. Status:Fully Paid",
    "Does not meet the credit policy. Status:Charged Off"
]

df_lc = df_lc[df_lc["loan_status"].isin(valid_status)]
```

```python
bad_status = [
    "Charged Off",
    "Late (31-120 days)",
    "Late (16-30 days)",
    "Default",
    "Does not meet the credit policy. Status:Charged Off"
]

df_lc["target"] = df_lc["loan_status"].apply(
    lambda x: 1 if x in bad_status else 0
)

df_lc.drop("loan_status", axis=1, inplace=True)

df_lc["target"].value_counts()
df_lc["target"].value_counts(normalize=True)
```

# Data Cleaning

```python
# Drop Rows with Critical Missing

df_lc = df_lc.dropna(subset=[
    "target",
    "int_rate",
    "loan_amnt",
    "term",
    "installment",
    "annual_inc",
    "fico_range_low",
    "fico_range_high",
    "dti",
])
```

```python
num_cols = [
    'revol_bal',
    'inq_last_6mths',
    'open_acc',
    'total_acc',
    'pub_rec',
    'delinq_2yrs',
    'revol_util'
]

df_lc[num_cols] = df_lc[num_cols].fillna(df_lc[num_cols].median())
```

```python
# Clean term
df_lc["term"] = df_lc["term"].str.extract(r'(\d+)').astype(int)
```

```python
# Create fico_score
df_lc["fico_score"] = (
    df_lc["fico_range_low"] + df_lc["fico_range_high"]
) / 2

df_lc.drop(["fico_range_low", "fico_range_high"], axis=1, inplace=True)
```

```python
# Convert earliest_cr_line tp credit_history_length
df_lc["credit_history_length"] = (
    pd.to_datetime("today") -
    pd.to_datetime(
        df_lc["earliest_cr_line"],
        format="%b-%Y",
        errors="coerce"
    )
).dt.days / 365

df_lc.drop("earliest_cr_line", axis=1, inplace=True)
```

```python
# Clean emp_length
df_lc["emp_length"] = (
    df_lc["emp_length"]
    .str.extract(r'(\d+)')
    .astype(float)
)

df_lc["emp_length"] = df_lc["emp_length"].fillna(-1)
```

```python
df_lc["inq_last_6mths"] = df_lc["inq_last_6mths"].fillna(0)
df_lc["delinq_2yrs"] = df_lc["delinq_2yrs"].fillna(0)
df_lc["total_acc"] = df_lc["total_acc"].fillna(df_lc["total_acc"].median())
```

```python
df_lc = df_lc.replace([float('inf'), -float('inf')], pd.NA)
```

```python
df_lc["credit_history_length"] = df_lc["credit_history_length"].fillna(
    df_lc["credit_history_length"].median()
)
```

```python
cat_cols = ['purpose','verification_status']
df_lc[cat_cols] = df_lc[cat_cols].fillna('Unknown')
```

# Feature Engineering

Create Core Engineered Features

```python
# Debt to income - How much income is used to repay loan
df_lc["dti_computed"] = (df_lc["installment"] * 12) / (df_lc["annual_inc"] + 1)
```

```python
# Loan to income - Size of loan relative to income
df_lc["loan_to_income"] = df_lc["loan_amnt"] / (df_lc["annual_inc"] + 1)
```

```python
# Normalize Credit Score - Keeps consistency with L&T later
df_lc["credit_score_norm"] = df_lc["fico_score"] / 900
```

```python
# Delinquency Indicator
df_lc["delinquency"] = df_lc["delinq_2yrs"]
df_lc["log_delinquency"] = np.log1p(df_lc["delinq_2yrs"])
df_lc.drop(["delinq_2yrs"], axis=1, inplace=True)
```

```python
# Log transformations
df_lc["log_loan_amount"] = np.log1p(df_lc["loan_amnt"])
df_lc["log_installment"] = np.log1p(df_lc["installment"])
df_lc["log_income"] = np.log1p(df_lc["annual_inc"].clip(
    upper=df_lc["annual_inc"].quantile(0.99)
))
df_lc["log_credit_history_length"] = np.log1p(df_lc["credit_history_length"])
df_lc["log_inquiries"] = np.log1p(df_lc["inq_last_6mths"])
df_lc["inquiries"] = df_lc["inq_last_6mths"]
df_lc.drop(["inq_last_6mths"], axis=1, inplace=True)
```
