Modifying data type of columns and filling empty values

```python
def convert_to_year(age):
    yrs = float(age.split("yrs")[0])
    mon = float(age.split("mon")[0].split()[-1])
    return yrs + mon/12
```

```python
df_lt["Employment.Type"] = df_lt["Employment.Type"].fillna("Unknown")

df_lt["AVERAGE.ACCT.AGE"] = df_lt["AVERAGE.ACCT.AGE"].apply(convert_to_year)
df_lt["AVERAGE.ACCT.AGE"] = df_lt["AVERAGE.ACCT.AGE"].astype(int)

df_lt["CREDIT.HISTORY.LENGTH"] = df_lt["CREDIT.HISTORY.LENGTH"].apply(convert_to_year)
df_lt["CREDIT.HISTORY.LENGTH"] = df_lt["CREDIT.HISTORY.LENGTH"].astype(int)

# categorical variables
df_lt["Employment.Type"] = df_lt["Employment.Type"].astype("category")
```

# Feature Engineering

Create Core Engineered Features

```python
# Normalize Loan-to-Value
df_lt["ltv_norm"] = df_lt["ltv"] / 100
```

```python
# Installment to Loan Ratio
df_lt["installment_ratio"] = df_lt["installment"] / df_lt["loan_amount"]
```

```python
# Loan Size vs Asset
df_lt["loan_to_asset"] = df_lt["loan_amount"] / df_lt["vehicle_value"]
```

```python
# Active Ratio
df_lt["active_ratio"] = df_lt["active_accounts"] / (df_lt["num_accounts"] + 1)
```

```python
# Overdue Ratio
df_lt["overdue_ratio"] = df_lt["overdue_accounts"] / (df_lt["num_accounts"] + 1)
```

```python
# Recent Credit Activity
df_lt["recent_ratio"] = df_lt["recent_accounts"] / (df_lt["num_accounts"] + 1)
```

```python
# Normalize Credit Score
df_lt["credit_score_norm"] = df_lt["credit_score"] / 900
```

```python
# Inquiry Intensity
df_lt["inquiry_intensity"] = df_lt["inquiries"] / (df_lt["num_accounts"] + 1)
```

```python
# Convert to binary
df_lt["has_delinquency"] = (df_lt["delinquency"] > 0).astype(int)
```

```python
df_lt["credit_history_length"] = df_lt["CREDIT.HISTORY.LENGTH"]
df_lt["avg_account_age"] = df_lt["AVERAGE.ACCT.AGE"]
```

```python
# Drop Raw Redundant Columns

df_lt.drop(columns=[
    "CREDIT.HISTORY.LENGTH",
    "AVERAGE.ACCT.AGE",
    "ltv",
], inplace=True)
```
